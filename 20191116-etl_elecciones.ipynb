Ahora, la columna `apoyo`.

```python
def get_support(value):
    if value.find("-") > 0:
        value_1 = "INDEPENDIENTE"
        value_2 = value[value.find("-") + 1:]
    else:
        value_1 = value
        value_2 = None
    return value_1, value_2
```

```python
base_total["partido_3"] = base_total.apply(lambda row: get_support(row.partido_2)[0], axis = 1)
base_total["apoyo"] = base_total.apply(lambda row: get_support(row.partido_2)[1], axis = 1)
```
